```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_circles
from sklearn.cluster import SpectralClustering, KMeans, DBSCAN
```

```python
# Example Clusters

X, clusters = make_circles(n_samples=1000, noise=.05, factor=.5, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
# K-means

km = KMeans(n_clusters=2, init='random', n_init = 10, max_iter=300)

predictions = km.fit_predict(X)
```

```python
plt.scatter(
    X[predictions == 0, 0], X[predictions == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[predictions == 1, 0], X[predictions == 1, 1],
    s=50, c='lightblue',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.grid()
plt.show()
```

```python
sp = SpectralClustering(n_clusters=2, affinity="nearest_neighbors")

sp_predictions = sp.fit_predict(X)
```

```python
plt.scatter(
    X[sp_predictions == 0, 0], X[sp_predictions == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[sp_predictions == 1, 0], X[sp_predictions == 1, 1],
    s=50, c='lightblue',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.grid()
plt.show()
```

```python
db = DBSCAN(eps=0.8, min_samples=200)

db_preds = db.fit_predict(X)
```
